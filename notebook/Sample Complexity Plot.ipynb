```python
def data_to_dicts(data, keys, stats_func):
    results = []
    for d in data:
        args = d["args"]
        entry = { k : getattr(args, k) if hasattr(args, k) else args[k] for k in keys } 
        # also value. 
        entry.update(stats_func(d["args"], d["stats"]))
        results.append(entry)
    return results

def extractor(args, stats):
    return [ stats[0][-1][i]["corr_eval"].max(dim=0)[0].mean() for i in range(4) ]

dicts = data_to_dicts(d, ["random_dataset_size"], extractor)
```
